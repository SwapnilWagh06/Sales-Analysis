### DATA PREPROCESSING   (CLEANING + NEW COLUMN IF REQUIRED)

```python
Sale=Sale.dropna(how='all')
Sale.head()
```

```python
Sale.shape  # so all row which contaion nan are drop from data set
```

## we will create new column which contain total price and the month column for monthly sale

```python
#Sale['Quantity Ordered']=Sale['Quantity Ordered'].astype('int32')
```

their is issue in some order id so we have to treat order id or date first

```python
Sale['Order Date'].nunique()
```

```python
Sale['Month'] = Sale['Order Date'].str[0:2]  ## to create new column for month
```

```python
#Sale['Month']=Sale['Month'].astype('int')

# ERROR COMES SHOW :- invalid literal for int() with base 10: 'Or'
#SO WE HAVE TO TREAT THE MONTH
```

```python
Sale = Sale[Sale['Order Date'].str[0:2]!='Or']
Sale.shape

## from not wual to or then all values comes whivh contaion non or value
```

```python
Sale['Month']=Sale['Month'].astype('int')  # now the codes run now we can treat price column also
```

```python
Sale['Quantity Ordered'] = pd.to_numeric(Sale['Quantity Ordered'])
Sale['Price Each'] = pd.to_numeric(Sale['Price Each'])
```

```python
Sale['Total_Price']=Sale['Quantity Ordered'].astype('int') * Sale['Price Each'].astype('float')
```

```python
Sale['Total_Price']=Sale['Total_Price'].astype('float')
```

```python
Sale.head()
```

### we will crate address column for proper result

```python
### have to write function for creating city

def city(address):
    return address.split(",")[1].strip(" ")
Sale['City']=Sale['Purchase Address'].apply(lambda x : city(x))
```

```python
Sale.head()
```

```python
Sale.City.unique()
```

```python
### we will follow same process for the state column 


def code(address):
    return address.split(",")[2].strip(" ")[0:2]
Sale['State']=Sale['Purchase Address'].apply(lambda x : code(x))
```

```python
Sale.head()
```

```python
Sale['Hrs'] = pd.to_datetime(Sale['Order Date']).dt.hour
Sale['Mins'] = pd.to_datetime(Sale['Order Date']).dt.minute
```

# LINEAR REGRESSION

```python
DF=Sale.copy()
```

```python
DF.head()
```

```python
DF.drop(['Order ID','Order Date','Purchase Address'],axis=1,inplace=True)
```

```python
DF.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lr=LabelEncoder()
```

```python
DF['Product']=lr.fit_transform(DF['Product'])
DF['City']=lr.fit_transform(DF['City'])
DF['State']=lr.fit_transform(DF['State'])
```

```python
### converted all categorical into numerical
```

```python
#DF=DF.drop(['Product'],axis=1)
```

```python
DF.head()
```

```python
DF.corr()
```

```python
sns.heatmap(DF.corr(),annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=DF.drop(['Total_Price','Price Each'],axis=1)
y=DF['Price Each']
```

```python
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.3,random_state=42 )
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
```

```python
LR.fit(X_train,y_train)
```

```python
LR.intercept_
```

```python
LR.coef_
```

```python
y_pred=LR.predict(X_test)
y_pred
```

```python
from  sklearn.metrics import mean_absolute_error

mean_absolute_error(y_test,y_pred)
```

```python
print("Testing Accuracy")
print(LR.score(X_test,y_test))
```
